```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/encoded-diabetic_data.csv")
df.head()
```

```python
df.info()
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df['readmitted_binary']=df['readmitted'].apply(lambda x:1 if x=='<30' else 0)
```

```python
df['total_cost']=df['num_medications']*120+df['time_in_hospital']*80
```

```python
threshold=0.5
df.dropna(thresh=len(df)*(1-threshold),axis=1,inplace=True)
```

```python
df.drop(columns=["encounter_id"],inplace=True)
```

```python
df.to_csv("cleaned_diabetic_data.csv",index=False,encoding='utf-8')
```
